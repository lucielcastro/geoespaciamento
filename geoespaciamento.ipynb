```python
#!pip install geopandas rtree fiona

import geopandas as gpd
import rtree
import pygeos
import mapclassify
import geobr
import folium
```

```python
#Abrindo nosso shp
gdf = geobr.read_municipality(code_muni='SP', year=2020) #municipios referentes ao Brasil
gdf.head()
```

```python
gua = gdf[gdf.name_muni=='Guarulhos']
gua
gua.crs='epsg:4674'
gua.crs
gua.explore()
gua.explore(
    color='gray',
    tooltip = 'name_muni',
    name = 'Guarulhos',
    tiles = 'CartoDB positron'
)
```

```python
gdf_base = geobr.read_schools(year=2020)
gdf_base.head()
```

```python
gdf_base.crs=='espg:4674'
gdf_base.crs
```

```python
gdf_base_gua = gpd.sjoin(gdf_base, gua)
gdf_base_gua.head()
```

```python
gdf_base_gua.columns
```

```python

m = gua.explore(
    color='gray',
    tooltip = 'name_muni',
    name = 'Guarulhos',

    tiles = 'CartoDB positron',
)
gdf_base_gua['centroid'] = gdf_base_gua['geometry'].centroid


gdf_base_gua.explore(
    m=m,
    column ='government_level',
    scheme = 'naturalbreaks',
    cmap ='Spectral',
    tooltip = False,
    name="Escolas",
    style_kwds=dict(color="black", weight=1, opacity= 0.4), #style argumen
    popup= ['name_school', 'location_type','address', 'government_level','phone_number',],
    legend=True,
    legend_kwds=dict(caption = 'Legenda'),
    marker_type="circle_marker",
    marker_kwds=dict(radius=5),
    height=600,
    width=600,
)
m
```
